```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import circuit_drawer, plot_bloch_multivector, plot_histogram
```

# Circuito 1: Capicúa (Vídeo)

```python
# Crear registros cuánticos y clásicos
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')

# Crear el circuito cuántico
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[2])
circuit.x(qreg_q[3])
circuit.cx(qreg_q[0], qreg_q[3])
circuit.cx(qreg_q[2], qreg_q[3])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3])
circuit.measure(qreg_q[0], creg_c[0])
circuit.measure(qreg_q[1], creg_c[1])
circuit.measure(qreg_q[2], creg_c[2])
circuit.measure(qreg_q[3], creg_c[3])

# Dibuja el circuito utilizando Matplotlib
circuit.draw(output='mpl')
```

```python
# Simular el circuito usando el simulador de estado cuántico
simulator = Aer.get_backend('statevector_simulator')

transpiled_circuit = transpile(circuit, simulator)
job = simulator.run(transpiled_circuit)
result = job.result()
statevector = result.get_statevector()

# Visualizar el vector de estado
plot_bloch_multivector(statevector)
```

```python
# Simular el circuito usando el simulador de conteo de frecuencias
simulator = Aer.get_backend('qasm_simulator')

transpiled_circuit = transpile(circuit, simulator)
job = simulator.run(transpiled_circuit)
qasm_result = job.result()

counts = qasm_result.get_counts()

# Visualizar el histograma de los resultados de las mediciones
plot_histogram(counts)
```
