```python
import pandas as pd
```

```python
df = pd.DataFrame()

# Ruta a los archivos Parquet
for i in range(6):
    path = 'dataset/parquet/000' + str(i) + '.parquet'

    # Leer el archivo Parquet en un DataFrame
    temp_df = pd.read_parquet(path, engine='pyarrow')
    df = pd.concat([df, temp_df])

# Mostrar el DataFrame
df = df.reset_index(drop=True)
# Copia de df
data = df.copy()
df
```

```python
#Para recuperar df desde la copia preprocesado
df = data.copy()
```

```python
# Arreglo de columnas inicial
df['title'] = df['title'].astype(str)
df['text'] = df['text'].astype(str)
df['identifier'] = df['identifier'].astype(int)
df['word_count'] = df['word_count'].astype(int)
df.drop(columns=['creator', 'publication_date', '__index_level_0__'], errors='ignore', inplace=True)
df.head()
```

```python
# Función para filtrar por palabras clave
def filter_by_keywords(df, category, keywords):
    # Crear una expresión regular que combine todas las palabras
    pattern_regex = '|'.join(keywords)
    
    # Buscar la subcadena en la columna 'title' utilizando la expresión regular
    result = df[df['title'].str.contains(pattern_regex, case=False, na=False)]
    result.loc[:,['category']]= category
    
    return result
```

```python
# Filtrar por palabras clave para ciencias sociales
df1 = filter_by_keywords(df, "Sociales", ["Derecho", "Política", "Político", "Economía", "Social"])
df1
```

```python
# Filtrar por palabras clave para ciencias puras
df2 = filter_by_keywords(df, "Ciencias", ["Ciencia", "Medicina", "Tecn", "Especies", "Animales", "Plantas", "Astronomía", "Programación", "Matemáticas", "Número"])
df2
```

```python
# Filtrar por palabras clave para religión
df3 = filter_by_keywords(df, "Religión", ["Religión", "Dios"])
df3
```

```python
# Filtrar por palabras clave para ciencias puras
df4 = filter_by_keywords(df, "Historia", ["Historia de", "Edad Media", "Edad Moderna", "Contempóranea"])
df4
```

```python
#Comprobar cantidad de cada categoría
df = pd.concat([df1, df2, df3, df4])
df['category'].value_counts()
```

```python
#Se eliminan duplicados por el identificador
counts = df['identifier'].value_counts()
unique_values = counts[counts == 1].index
df = df[df['identifier'].isin(unique_values)].drop(["identifier"], axis=1)
df['category'].value_counts()
```

```python
# Se guardan los textos filtrados
df.to_csv('dataset/parquet/df_parquet_full.csv', index=False)
```
